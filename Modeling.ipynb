# Import Datasets

```python
import pandas as pd
```

```python
content_df_new = pd.read_csv('ContentCleaned.csv')
reactions_df_new = pd.read_csv('ReactionsCleaned.csv')
reaction_types_df_new = pd.read_csv('ReactionTypesCleaned.csv')
```

```python
content_df_new.head()
```

Let's start merging them using common columns - Content ID and Reaction Type

```python
merged_df = pd.merge(content_df_new, reactions_df_new, on='Content ID')
```

```python
merged_df.head()
```

```python
reaction_types_df_new
```

```python
final_df = pd.merge(merged_df, reaction_types_df_new, on='Reaction Type')
```

```python
final_df
```

# Finding the top 5 performing category

```python
top_5_categories = final_df.groupby('Category')['Score'].sum().nlargest(5)
```

```python
top_5_categories
```

```python
content_type_counts = final_df['Content Type'].value_counts().nlargest(5)
```

```python
content_type_counts
```

```python
content_type_counts = final_df['Reaction Type'].value_counts().nlargest(5)
```

```python
content_type_counts
```

```python
final_df.to_csv('FinalDataset.csv', index=False)
```
